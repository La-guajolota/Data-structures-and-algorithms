* EXAMPLE:n^2 + n = O(n^2) 

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
inputs = np.arange(25)

outputs1 = inputs
outputs2 = inputs**2
outputs3 = outputs1 + outputs2

plt.plot(inputs, outputs1, label='n')
plt.plot(inputs, outputs2, label='n^2')
plt.plot(inputs, outputs3, label='O(n^2+n)')
plt.legend()

plt.xlabel('Inputs size')
plt.ylabel('Growth rate')
plt.title('Example: n^2 + n = O(n^2)')

plt.grid()
plt.xticks(inputs)

plt.show()
```

# Omega Notetion

* Definition: “f(n) is omega of g(n).” or f(n) = Ω(g(n)) if there are two +ve
constants c and n0 such that
c g(n) ≤ f(n) for all n ≥ n0

* Find relationship of f(n) = n^

F(n) = Ω (g(n))

```python
inputs = np.arange(11)
c = 2 #Positive constant

outputs1 = inputs**c # f(n) = n^c
outputs2 = c**inputs # g(n) = c^n

plt.plot(inputs, outputs1, label='f(n) = n^c')
plt.plot(inputs, outputs2, label='g(n) = c^n')

plt.legend()

plt.xlabel('Inputs size')
plt.ylabel('Growth rate')
plt.title('F(n) = Ω(g(n))')

plt.grid()
plt.xticks(inputs)

plt.show()
```

```python
n = symbols('n')

f = n**c
g = c**n
ng = n*g

# derivates
derf = diff(f, n)
derg = diff(g, n)
derng = diff(ng, n)


print("´f(x) =", derf)
print("´g(x) =", derg)
print("´(ng(x)) =", derng)
```

1. f(n) has a exponenial growth rate
2. g(n) has the same exponenial growth rate as f(n) times log(2)
3. ng(n) is to g(n) as g(n) is to f(n), but it is faster by a constant exponential growth of 2**n

```python
import numpy as np
import matplotlib.pyplot as plt

# Define el rango de valores de n
n_values = np.linspace(1, 5)

# Complejidad O(1)
complexity_O1 = np.ones_like(n_values)

# Complejidad O(log n)
complexity_Ologn = np.log10(n_values)

# Complejidad O(n)
complexity_On = n_values

# Complejidad O(n log n)
complexity_Onlogn = n_values * np.log10(n_values)

# Complejidad O(n^2)
complexity_On2 = n_values**2

# Complejidad O(2^n)
complexity_O2n = 2**n_values

# Graficar las complejidades
plt.figure(figsize=(10, 6))

plt.plot(n_values, complexity_O1, label='O(1)')
plt.plot(n_values, complexity_Ologn, label='O(log n)')
plt.plot(n_values, complexity_On, label='O(n)')
plt.plot(n_values, complexity_Onlogn, label='O(n log n)')
plt.plot(n_values, complexity_On2, label='O(n^2)')
plt.plot(n_values, complexity_O2n, label='O(2^n)')

plt.xlabel('n')
plt.ylabel('Complexity')
plt.title('Time Complexity Notations')
plt.legend()
plt.grid(True)
plt.show()
```
